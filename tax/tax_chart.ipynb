```python
import matplotlib.pyplot as plt


principal = [item*10000 for item in range(3,16)]

def calculate_tax(P):
    tax = 0
    if P>24000:
        if P<32333:
            tax = (P-24000)*.25
        else:  
            tax = (8333*0.25) + ((P-32333)*0.30)
    return tax        

taxes = [calculate_tax(n) for n in principal]
take_home = [(p - calculate_tax(p)) for p in principal]
print(principal)
print(taxes)
print(take_home)
```

```python
# Percentages
import numpy as np

divided_tax = list(np.divide(taxes,principal))
percentageTax = list(np.array(divided_tax)*100)
divided_take_home = list(np.divide(take_home,principal))
percentage_take_home = list(np.array(divided_take_home)*100)

print(divided_tax)
print(percentageTax)
print(divided_take_home)
print(percentage_take_home)
```

```python
plt.figure(figsize=(30, 12))
# Values
plt.subplot(121)
plt.grid()
plt.plot(principal,taxes, 'r--')
plt.plot(principal,take_home, 'g-')
plt.plot(principal,take_home,'go')
plt.axis([0,170000,0,120000])
plt.ylabel('PAYE & Take Home')
plt.xlabel('Income Earned')

# Percentages
plt.subplot(122)
plt.grid()
plt.plot(principal,percentageTax,'r--')
plt.plot(principal,percentage_take_home)
plt.axis([0,170000,0,100])
plt.ylabel('Percentage')
plt.xlabel('Income Earned')
```
